```python
import os
from elasticsearch import Elasticsearch

# Set OpenAI API key
os.environ["OPENAI_API_KEY"] = "sk..........." #removed my openapi key

# Elasticsearch Cloud credentials
cloud_id = "4f49....." #removed my cloud id
es_user = "elastic"
es_pass = "My-Password" # removed my password

# Connect to Elasticsearch Cloud
es_client = Elasticsearch(
    cloud_id=cloud_id,
    basic_auth=(es_user, es_pass)
)
```

```python
from langchain_community.document_loaders import DirectoryLoader
from langchain.text_splitter import CharacterTextSplitter

loader = DirectoryLoader("data", glob="*.txt")
docs = loader.load()

splitter = CharacterTextSplitter(chunk_size=500, chunk_overlap=50)
chunks = splitter.split_documents(docs)

print(f"📚 Total chunks: {len(chunks)}")
```

```python
from langchain_community.vectorstores.elasticsearch import ElasticsearchStore
from langchain_huggingface import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# Create index
index_name = "rag-colab-index"

# Remove old index if exists
if es_client.indices.exists(index=index_name):
    es_client.indices.delete(index=index_name)

# Index the documents
store = ElasticsearchStore.from_documents(
    chunks,
    embedding=embeddings,
    es_connection=es_client,
    index_name=index_name
)
```

```python
from langchain_openai import ChatOpenAI
from langchain.chains import RetrievalQA

llm = ChatOpenAI(temperature=0.7, model_name="gpt-3.5-turbo")

qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    retriever=store.as_retriever(search_kwargs={"k": 5}),
    return_source_documents=True
)

# Query
query = "How can AI help in patient health monitoring?"
result = qa_chain.invoke(query)

print(" Answer:\n", result['result'])

print("\nSource Documents:")
for doc in result['source_documents']:
    print(f"- {doc.metadata.get('source', 'Unknown')}")
```
